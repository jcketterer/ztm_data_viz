```python
# Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.graph_objects as go
```

```python
# Load the data
df = pd.read_csv("happiness.csv")
df.head()
```

```python
# Information about the df
df.info()
```

# Bar Chart - Horizontal - Rank the Happiest countries

```python
# Defining the plot size
fig, ax = plt.subplots(figsize = (12, 4))

# Retrieve the top 10 countries
data = df[['Country name',
           'happiness']].nlargest(10, 'happiness').sort_values(by = 'happiness',
                                                               ascending = True)
y_pos = np.arange(len(data['Country name']))

# Bar chart
plt.barh(y = y_pos,
         width = data.happiness)

# add countries
plt.yticks(ticks = y_pos,
           labels = data['Country name'])

# Change the font size
ax.tick_params(axis = "both", labelsize = 15)

# Add a title
fig.suptitle('Happiness of the top 10 countries, ranked', fontsize = 15)

# Plotting
plt.show()
```

# Vertical bar plot

```python
# Defining the plot size
fig, ax = plt.subplots(figsize = (12, 4))

# Retrieve the top 10 countries
data = df[['Country name',
           'happiness']].nlargest(10, 'happiness').sort_values(by = 'happiness',
                                                               ascending = True)
x_pos = np.arange(len(data['Country name']))

# Create a color list
my_color = np.where(data['Country name'] == 'Norway', 'orange', 'skyblue')

# Bar chart
plt.bar(x = x_pos,
        height = data.happiness,
        color = my_color)

# add countries
plt.xticks(ticks = y_pos,
           labels = data['Country name'],
           rotation = 45
           )

# Change the font size
ax.tick_params(axis = "both",
               labelsize = 15)

# Add a title
fig.suptitle('Happiness of the top 10 countries, ranked',
             fontsize = 15)

# Plotting
plt.show()
```

# Lollipop chart

```python
# Defining the plot size
fig, ax = plt.subplots(figsize = (12, 4))

# Retrieve the top 10 countries
data = df[['Country name',
           'Generosity']].nlargest(10, 'Generosity').sort_values(by = 'Generosity',
                                                               ascending = True)
x_pos = np.arange(len(data['Country name']))

# Lollipop
plt.stem(data.Generosity,
         linefmt = 'skyblue',
         markerfmt = 'orange')

# add countries
plt.xticks(ticks = y_pos,
           labels = data['Country name'],
           rotation = 45
           )

# Change the font size
ax.tick_params(axis = "both",
               labelsize = 15)

# Add a title
fig.suptitle('Generosity of the top 10 countries, ranked',
             fontsize = 15)

# X and Y labels
plt.xlabel('Countries', fontsize = 15)
plt.ylabel('Generosity Score', fontsize = 15)

# Plotting
plt.show()
```

# Spider Charts

```python
# Look at the data
df.head()
```

```python
# Import functions
from sklearn import preprocessing
from math import pi
```

```python
# Normalize with min max scaler
min_max_scaler = preprocessing.MinMaxScaler()
df_scaled = df
df_scaled.iloc[:, 1:] = min_max_scaler.fit_transform(df.iloc[:,1:])
df_scaled.head()
```

```python
# Categories
categories = list(df_scaled)[1:]
N = len(categories)
```

```python
# Compute the position of the categories
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
angles
```

```python
# Pick countries
A = "Germany"
B = "Portugal"
```

```python
# Defining the plot size
ax = plt.subplot(111, polar = True)

# add something to xticks
plt.xticks(ticks = angles[:-1],
           labels = categories,
           rotation = 0,
           color = 'black',
           size = 12
           )

# add something to yticks
ax.set_rlabel_position(0)
plt.yticks([0.3, 0.7],
           labels = ["0.3", "0.7"],
           color = "grey",
           size = 7)
plt.ylim(0,1)


# Element A
values = df_scaled.loc[df_scaled['Country name'] == A].drop(
    columns = 'Country name')
values = values.values.flatten().tolist()
values += values[:1]

# Spider Chart
ax.plot(angles,
        values,
        label = A,
        linewidth = 2,
        linestyle = 'solid')

ax.fill(angles,
        values,
        'b',
        alpha = 0.1)

# Element B
values = df_scaled.loc[df_scaled['Country name'] == B].drop(
    columns = 'Country name')
values = values.values.flatten().tolist()
values += values[:1]

# Spider Chart
ax.plot(angles,
        values,
        label = B,
        linewidth = 2,
        linestyle = 'solid')

ax.fill(angles,
        values,
        'r',
        alpha = 0.1)

# Plotting
plt.legend(loc = 'upper right', bbox_to_anchor = (0.1, 0.1))
plt.show()
```
